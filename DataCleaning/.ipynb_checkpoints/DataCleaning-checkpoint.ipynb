# Data Cleaning Pipeline
In this notebook we will go over how to apply our data cleaning algorithm to data downloaded from the chandra archive. To run this notebook, you need just to have ciao install (https://cxc.cfa.harvard.edu/ciao/download/) and an ObsID downloaded from chaser (https://cda.harvard.edu/chaser/). This notebook will use code we have written specifically tailored to galaxy clusters. When you install ciao be sure to include `acis_bkg_evt`.

There are a few additional python packages you will need. I suggest you use the requirements.txt file to install the additional modules. Otherwise, I enumerate them here:
 -  easygui
 - deproject
 - tqdm

```python
#------------------------------------IMPORTS-----------------------------------#
#----------------------------------GENERAL IMPORTS-----------------------------#
import os
import sys
import shutil
import easygui as gui
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from ciao_contrib.runtool import *
#-------------------------------Additional IMPORTS----------------------------#
from Misc.read_input import read_input_file
from Misc.Bkg_sub import run_bkg_sub, create_clean_img, exp_corr, create_clean_img_merge
from Preliminary.unzip import unzip
from Preliminary.Merge import merge_objects
from Preliminary.CCD_split import split_ccds
from Preliminary.FaintCleaning import FaintCleaning
from Preliminary.CreateLightcurves import bkg_clean_srcs, bkg_lightcurve
from Preliminary.chips_ccd import display_ccds, display_entire
```

## Define the parameters for cleaning the data

```python
home_dir = '/home/carterrhea/Documents/Perseus'  # Path to ObsID (not including the ObsID)
dir_list = ['4289']  # List of ObsIDs (ex. ['1111', '1112'])
name = 'Perseus'  # Name of cluster -- this is for the output
```

## Run the code
In this next section we will use the parameters defined above and run the script to clean the data. The clean data will be in home_dir/dir_list/repro. So if you set home_dir='/home/Documents' and dir_list='1111', you will find the reprocessed data under '/home/Documents/1111/repro'.

The code requires you to select the background CCD -- you will be given a prompt and asked to choose.

```python
#---------------------------Read in data----------------------------------#
print("Running Preliminary Steps...")
print("#-------STARTING ANALYSIS ON %s-------#"%name)
#inputs,merge_bool = read_input_file(input_file)
os.chdir(home_dir)
#Unzip all relavent files
unzip(home_dir,dir_list)
print("Generating fits and image files for each individual CCD...")
ccds = split_ccds(home_dir,dir_list)
#if not os.path.exists(home_dir+'/'+name):
#    os.makedirs(home_dir+'/'+name)
#We must clean each observation first :)
print("Beginning cleaning process for each individual obsid...")
for obsid_ in dir_list: #left as a list to keep input deck the same and sample :)
    os.chdir(home_dir + '/' + obsid_ + '/Background')
    print("We are on obsid %s"%obsid_)
    print("    Now let us pick our background ccd...")
    #Lets take a look at each ccd and pick our background and src ccds
    bkg_ccd = display_ccds(ccds,obsid_,Merge=True)
    print("The background CCD chosen is CCD#%s\n"%bkg_ccd[-1])
    print("    We can now create a lightcurve for the background...")
    bkg_clean_srcs(bkg_ccd)
    bkg_lightcurve(bkg_ccd,obsid_)
    print("    We need to clean our diffuse emission...")
    filenames = FaintCleaning(home_dir,obsid_,bkg_ccd,0,0,ccds[obsid_])
    #We have to create bkg-subtracted images for each obsid because we need them for our merged image!
    os.chdir(home_dir + '/' + obsid_ + '/repro')
    print("    Creating Clean Image...")
    create_clean_img_merge(filenames)
    print("    Running Background Subtraction...")
    run_bkg_sub(filenames['evt2_repro_uncontam'], filenames['evt_uncontam_img'], obsid_, filenames)
print("Beginning Merged Calculations...")
print("    Merging obsids...")
os.chdir(home_dir)
merge_objects(dir_list, name, clean='yes')
```
